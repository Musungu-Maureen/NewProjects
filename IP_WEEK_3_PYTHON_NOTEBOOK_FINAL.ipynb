```python
#importing the libraries we will be using 
import pandas as pd
```

```python
#loading our datasets 
df1 = pd.read_csv('Telcom_dataset.csv')
df1
#cleaning this dataframe
#taking down the none-useful columns
df1.drop('CELL_ON_SITE', axis=1, inplace=True)
df1.drop('COUNTRY_A', axis=1, inplace=True)
df1.drop('COUNTRY_B', axis=1, inplace=True)
df1

#then we correct the misspelt words
df1.columns.values[0:1] = ['PRODUCT']
df1.columns.values[2:3] = ['DATE_TIME']

#Checking for null values and checking their sum
df1.isnull().values.any()
df1.isnull().sum()

#the column 'SITE_ID' contains 716 null values
#we shall substitute all the null values with a "."
df1 = df1.fillna('.')
df1

#checking for duplicates
df1.duplicated().values.sum()
#we see there are 82 duplicated records and we therefore drop them
df1.drop_duplicates()
df1
```

```python
df2 = pd.read_csv('Telcom_dataset2.csv')
df2
#cleaning this dataframe
#taking down the none-useful columns
df2.drop('CELL_ON_SITE', axis=1, inplace=True)
df2.drop('COUNTRY_A', axis=1, inplace=True)
df2.drop('COUNTRY_B', axis=1, inplace=True)
df2

#we shall then change the names of the columns to make them similar to the first dataframe column names
df2.columns.values[3:5] = ['DW_A_NUMBER_INT','DW_B_NUMBER_INT']
df2

#Checking for missing values
df2.isnull().values.any()

#since the result is true, we go ahead to get how many they are
df2.isnull().sum()

#The Column 'SITE_ID' contains 676 null values. we shall fill these values with "."
df2.fillna('.')

#Checking for duplicates 
df2.duplicated()

#We shall drop the duplicated value which is present in the 5000th record
df2.drop_duplicates()
df2
```

```python
df3= pd.read_csv('Telcom_dataset3.csv')
df3

#dropping the data we shall not be using 
df3.drop('CELL_ON_SITE', axis=1, inplace=True)
df3.drop('COUNTRY_A', axis=1, inplace=True)
df3.drop('COUNTRY_B', axis=1, inplace=True)
df3

#Correcting the column names 
df3.columns.values[5:7] = ['CELL_ID','SITE_ID']
df3

#Checking for missing values
df3.isnull().values.any()

#we count the number of null values
df3.isnull().sum()

#the 'SITE_ID' column contains 605 null values. We Replace them with "."
df3.fillna(".")

#Checking for duplicates 
df2.duplicated()

#We shall drop the duplicated value which is present in the 5000th record
df3.drop_duplicates()
df3
```

```python
#Loading the cells_geo dataset. 
#Using the delimiter ';' to split the data and align the rows into their corresponding columns. 
df_geo= pd.read_csv('cells_geo.csv', encoding='utf8', index_col=0, delimiter = ';')
df_geo

#checking for null values
df_geo.isnull().values.any()

#Getting the total null values
df_geo.isnull().values.sum()

#Replacing all the null values with '.'
df_geo.fillna(".")

#Checking for duplicates
df_geo.duplicated()

#There are no duplicates in this dataset 
#We therefore load our dataset
df_geo
```

```python
#Merging the Telecom Datasets together 
df_all = pd.concat((df1, df2, df3), axis=0, ignore_index=True)
df_all
```

```python
#Merging the telecom datasets to the cells_geo dataset
```

```python
final_df = pd.merge(df_all, df_geo, how = 'inner', left_on = 'SITE_ID', right_on = 'SITE_CODE')
final_df
```
